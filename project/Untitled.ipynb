```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'combined_salaries.csv'
data = pd.read_csv(file_path)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

sns.boxplot(ax=axes[0], data=data, x='Salary', y='Experience Level')
axes[0].set_title('Salary Distribution by Experience Level')
axes[0].set_xlabel('Salary')
axes[0].set_ylabel('Experience Level')

avg_salary_exp = data.groupby('Experience Level')['Salary'].mean().sort_values()
axes[1].barh(avg_salary_exp.index, avg_salary_exp.values)
axes[1].set_title('Average Salary by Experience Level')
axes[1].set_xlabel('Average Salary')
axes[1].set_ylabel('Experience Level')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.boxplot(data=data, x='Salary', y='Experience Level', showfliers=False)
plt.title('Salary Distribution by Experience Level (Excluding Outliers)')
plt.xlabel('Salary')
plt.ylabel('Experience Level')
plt.show()
```

```python
experience_mapping = {'EN': 0, 'MI': 1, 'SE': 2, 'EX': 3}
data['Experience Level Code'] = data['Experience Level'].map(experience_mapping)

correlation_exp_salary = data['Experience Level Code'].corr(data['Salary'])

correlation_exp_salary
```
