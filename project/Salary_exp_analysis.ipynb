```python
job_titles = ['Data Scientist', 'Data Engineer', 'Data Analyst', 'Machine Learning Engineer', 
              'Software Engineer', 'Data Manager', 'Business Analyst', 'AI Architect', 
              'DevOps Engineer', 'Director', 'Marketing Specialist', 'Project Manager', 
              'Consultant', 'Data Architect', 'Sales', 'Operations', 'Finance', 'Specialist']

# Filter the DataFrame
filtered_df = df[df['Job Title'].isin(job_titles)]
```

```python
filtered_df
```

```python
salary_summary = filtered_df.groupby('Job Title')['Salary'].agg(['mean', 'median', 'min', 'max', 'count'])
salary_summary['range'] = salary_summary['max'] - salary_summary['min']

salary_summary
```

```python
exp_summary = filtered_df.groupby('Experience Level')['Salary'].agg(['mean', 'median', 'min', 'max', 'count'])
exp_summary['range'] = exp_summary['max'] - exp_summary['min']

exp_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

sns.boxplot(ax=axes[0], data=filtered_df, x='Salary', y='Experience Level')
axes[0].set_title('Salary Distribution by Experience Level')
axes[0].set_xlabel('Salary')
axes[0].set_ylabel('Experience Level')

avg_salary_exp = filtered_df.groupby('Experience Level')['Salary'].mean().sort_values()
axes[1].barh(avg_salary_exp.index, avg_salary_exp.values)
axes[1].set_title('Average Salary by Experience Level')
axes[1].set_xlabel('Average Salary')
axes[1].set_ylabel('Experience Level')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.boxplot(data=filtered_df, x='Salary', y='Experience Level', showfliers=False)
plt.title('Salary Distribution by Experience Level (Excluding Outliers)')
plt.xlabel('Salary')
plt.ylabel('Experience Level')
plt.show()
```
