```python
import pandas as pd
from datetime import timedelta

# Load the data
df = pd.read_csv('billboard.csv', encoding='latin1')
```

```python
suffixes = ["st", "nd", "rd"] + ["th"] * 7 
value_vars = [f'x{i+1}{suffixes[i%10]}.week' for i in range(76)]

value_vars[10] = 'x11th.week'
value_vars[11] = 'x12th.week'
value_vars[12] = 'x13th.week'
```

```python
df_long = df.melt(id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
                  var_name='week', value_name='rank', 
                  value_vars=value_vars)

df_long['week'] = df_long['week'].str.extract('(\d+)').astype(int)
df_long['date.entered'] = pd.to_datetime(df_long['date.entered'])
df_long['date'] = df_long.apply(lambda row: row['date.entered'] + timedelta(weeks=row['week']-1), axis=1)
df_long.drop(columns='date.entered', inplace=True)
```

```python
df_long = df_long[['year', 'artist', 'track', 'time', 'date', 'week', 'rank']]

part1_df = df_long

part1_df = part1_df.sort_values(by=['artist', 'week'])
part1_df
```
